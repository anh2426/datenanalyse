```python
# Anzahl Radfahrer 2022

file_path = "data/rad_2022_tage_export_09_02_23_r.csv"

# DataFrame erstellen
df = pd.read_csv(file_path)

gesamt_summe_2022 = df['gesamt'].sum()

# Datumsfeld in ein DateTime-Objekt umwandeln
df['datum'] = pd.to_datetime(df['datum'])

# Datumsfeld als Jahr extrahieren
df['jahr'] = df['datum'].dt.year

# Daten für das Jahr 2022 filtern
#df_2022 = df[df['jahr'] == 2022]

# Summe der Spalte 'gesamt' für das Jahr 2022 berechnen
#gesamt_summe_2022 = df_2022['gesamt'].sum()

# Altair-Diagramm erstellen
chart = alt.Chart(pd.DataFrame({'Jahr': ['2022'], 'Gesamtsumme': [gesamt_summe_2022]})).mark_bar().encode(
    x='Jahr:O',
    y='Gesamtsumme:Q',
    tooltip=['Jahr', 'Gesamtsumme']
).properties(
    title='Anazhl der Radfahrer für das Jahr 2022',
    width=200,
    height=150
)

print(gesamt_summe_2022)
chart
```

```python
# Funktion: Automatisches Erstellen eines Diagramms aus allen Dateipfaden
def create_chart(file_paths):
    # Gesamtsummen aus den CSV-Dateien berechnen
    gesamt_summen = [pd.read_csv(file_path)['gesamt'].sum() for file_path in file_paths]

    # Jahre aus der jeweiligen Spalte 'datum' extrahieren
    jahre = [pd.to_datetime(pd.read_csv(file_path)['datum']).dt.year.iloc[0] for file_path in file_paths]
    
    #Jahre aus dem Dateipfad extrahieren
    #jahre = [str(pd.to_datetime(file_path.split('_')[1], format='%Y').strftime('%Y')) for file_path in file_paths]
    
    # Altair-Diagramm erstellen
    chart = alt.Chart(pd.DataFrame({'Jahr': jahre, 'Anzahl Radfahrer': gesamt_summen})).mark_bar().encode(
        alt.X('Jahr:O'), # O, because time is ordinal
        alt.Y('Anzahl Radfahrer:Q'), #scale does nothing here # Q because the sum is quantitative
        tooltip=['Jahr', 'Anzahl Radfahrer']  # tooltip = text to show, when user's cursor is over a mark
    ).properties(
        title='Anzahl Radfahrer pro Jahr von 2008 bis 2022',
        width=400,
        height=150
    )

    return chart

# Diagramm erstellen
chart = create_chart(file_paths)

# Diagramm anzeigen
chart
```

```python
import altair as alt
# Anzahl Radfahrer pro Monat in 2022
file_path_2022 = "data/rad_2022_tage_export_09_02_23_r.csv"

# DataFrame für 2022 erstellen
df_2022 = pd.read_csv(file_path_2022)

# 'datum' in ein DateTime-Objekt umwandeln
df_2022['datum'] = pd.to_datetime(df_2022['datum'])

# Monat aus dem Datum extrahieren
df_2022['monat'] = df_2022['datum'].dt.strftime('%B')

# Gesamtsummen pro Monat berechnen
gesamtsummen_pro_monat = df_2022.groupby('monat')['gesamt'].sum().reset_index()

# Reihenfolge der Monate festlegen
monatsreihenfolge = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

# Altair-Diagramm erstellen
chart = alt.Chart(gesamtsummen_pro_monat).mark_bar().encode(
    x=alt.X('monat:N', sort=monatsreihenfolge),
    y='gesamt:Q',
    tooltip=['monat:N', 'gesamt:Q']
).properties(
    title='Anzahl Radfahrer pro Monat in 2022',
    width=400,
    height=300
)

# Diagramm anzeigen
chart
```

```python
# Vergleich der Monatsdaten

file_path_2022 = "data/rad_2022_tage_export_09_02_23_r.csv"
file_path_2021 = "data/rad_2021_tage_export_09_02_23_r.csv"

# DataFrames für 2022 und 2021 erstellen
df_2022 = pd.read_csv(file_path_2022)
df_2021 = pd.read_csv(file_path_2021)

# 'datum' in ein DateTime-Objekt umwandeln
df_2022['datum'] = pd.to_datetime(df_2022['datum'])
df_2021['datum'] = pd.to_datetime(df_2021['datum'])

# Monatsnamen aus dem Datum extrahieren
df_2022['monat'] = df_2022['datum'].dt.strftime('%B')
df_2021['monat'] = df_2021['datum'].dt.strftime('%B')

# Gesamtsummen pro Monat und Jahr berechnen
gesamtsummen_2022 = df_2022.groupby(['monat', df_2022['datum'].dt.year])['gesamt'].sum().reset_index()
gesamtsummen_2021 = df_2021.groupby(['monat', df_2021['datum'].dt.year])['gesamt'].sum().reset_index()

# Altair-Diagramm erstellen
chart = alt.Chart(pd.concat([gesamtsummen_2022, gesamtsummen_2021])).mark_bar().encode(
    x=alt.X('monat:N', sort=monatsreihenfolge),
    y='gesamt:Q',
    color='datum:N',
    column='datum:N',
    tooltip=['datum:N', 'monat:N', 'gesamt:Q']
).properties(
    title='Anzahl der Radfahrer pro Monat in verschiedenen Jahren',
    width=100,
    height=300
)

# Diagramm anzeigen
chart
```

```python
def create_chart(file_paths):
    # Gesamtsummen aus den CSV-Dateien berechnen
    gesamt_summen = [pd.read_csv(file_path)['gesamt'].sum() for file_path in file_paths]

    # Anzahl der Einträge pro Zählstelle für jede Datei berechnen
    eintraege_pro_datei = [pd.read_csv(file_path)['zaehlstelle'].nunique() for file_path in file_paths]

    # Jahre aus der jeweiligen Spalte 'datum' extrahieren
    jahre = [pd.to_datetime(pd.read_csv(file_path)['datum']).dt.year.iloc[0] for file_path in file_paths]
    
    # Altair-Diagramm erstellen
    chart = alt.Chart(pd.DataFrame({'Jahr': jahre, 'Anzahl Radfahrer': gesamt_summen, 'Einträge pro Zählstelle': eintraege_pro_datei})).mark_bar().encode(
        alt.X('Jahr:O'),  # O, because time is ordinal
        alt.Y('Anzahl Radfahrer:Q'),  # Q because the sum is quantitative
        tooltip=['Jahr', 'Anzahl Radfahrer', 'Einträge pro Zählstelle']  # tooltip = text to show when the cursor is over a mark
    ).properties(
        title='Anzahl Radfahrer pro Jahr von 2008 bis 2022',
        width=400,
        height=150
    )

    return chart

# Diagramm erstellen
chart = create_chart(file_paths)

# Diagramm anzeigen
chart
```
