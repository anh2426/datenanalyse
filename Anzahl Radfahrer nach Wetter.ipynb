# Werte von 2017 bis 2022

```python
def create_chart(file_paths):
    # Daten für alle Jahre kombinieren
    data = pd.concat([pd.read_csv(file_path) for file_path in file_paths])

    # Konvertiere 'max.temp' zu numerischem Datentyp
    data['max.temp'] = pd.to_numeric(data['max.temp'], errors='coerce')

    # Bins für die Temperatur erstellen
    temperature_bins = [-float('inf'), -5, 0, 5, 10, 15, 20, 25, 30, float('inf')]
    temperature_labels = ['-∞ to -5', '-5 to 0', '0 to 5', '5 to 10', '10 to 15', '15 to 20', '20 to 25', '25 to 30', '30+']

    # Daten in Bins einteilen
    data['Temperature Bin'] = pd.cut(data['max.temp'], bins=temperature_bins, labels=temperature_labels, right=False)

    # Anzahl der Radfahrer pro Temperatur-Bin und Jahr berechnen
    result = data.groupby(['Temperature Bin', pd.to_datetime(data['datum']).dt.year])['gesamt'].sum().reset_index()

    # Altair-Diagramm erstellen
    chart = alt.Chart(result).mark_bar().encode(
        alt.X('Temperature Bin:N', title='Temperaturbereich in °C'),
        alt.Y('gesamt:Q', title='Anzahl Radfahrer'),
        color=alt.Color('Temperature Bin:N', legend=None),
        column=alt.Column('datum:O', title='Jahr'),
        tooltip=['Temperature Bin:N', 'gesamt:Q']
    ).properties(
        title='Anzahl Radfahrer für verschiedene Temperaturbereiche von 2017 bis 2022',
        width=100,
        height=300
    )

    return chart

chart = create_chart(file_paths)

chart
```

# Aufkummulierte Werte von 2017 bis 2022

```python
def create_chart(file_paths):
    # Daten für alle Jahre kombinieren
    data = pd.concat([pd.read_csv(file_path) for file_path in file_paths])

    # Konvertiere 'max.temp' zu numerischem Datentyp
    data['max.temp'] = pd.to_numeric(data['max.temp'], errors='coerce')

    # Bins für die Temperatur erstellen
    temperature_bins = [-float('inf'), -5, 0, 5, 10, 15, 20, 25, 30, float('inf')]
    temperature_labels = ['-∞ to -5', '-5 to 0', '0 to 5', '5 to 10', '10 to 15', '15 to 20', '20 to 25', '25 to 30', '30+']

    # Daten in Bins einteilen
    data['Temperature Bin'] = pd.cut(data['max.temp'], bins=temperature_bins, labels=temperature_labels, right=False)

    # Anzahl der Radfahrer pro Temperatur-Bin und Jahr berechnen
    result = data.groupby(['Temperature Bin', pd.to_datetime(data['datum']).dt.year])['gesamt'].sum().reset_index()

    # Kumulierte Summen pro Temperatur-Bin berechnen
    result['Cumulative Total'] = result.groupby('Temperature Bin')['gesamt'].cumsum()

    # Altair-Diagramm erstellen
    chart = alt.Chart(result).mark_bar().encode(
        alt.X('Temperature Bin:N', title='Temperatur-Bereich (°C)', sort=temperature_labels),
        alt.Y('Cumulative Total:Q', title='Kumulierte Anzahl Radfahrer'),
        color=alt.Color('Temperature Bin:N', legend=None),
        column=alt.Column('datum:O', title='Jahr'),
        tooltip=['Temperature Bin:N', 'Cumulative Total:Q']
    ).properties(
        title='Kumulierte Anzahl Radfahrer für verschiedene Temperaturbereiche von 2008 bis 2022',
        width=100,
        height=300
    )

    return chart

chart = create_chart(file_paths)

chart
```

# Anzahl Radfahrer pro Temperaturbereich von 2019 bis 2022

```python
def create_charts(file_paths):
    # Daten für alle Jahre kombinieren
    data = pd.concat([pd.read_csv(file_path) for file_path in file_paths])

    # Konvertiere 'max.temp' zu numerischem Datentyp
    data['max.temp'] = pd.to_numeric(data['max.temp'], errors='coerce')

    # Bins für die Temperatur erstellen
    temperature_bins = [-float('inf'), -5, 0, 5, 10, 15, 20, 25, 30, float('inf')]
    temperature_labels = ['-∞ bis -5','-5 bis 0', '0 bis 5', '5 bis 10', '10 bis 15', '15 bis 20', '20 bis 25', '25 bis 30', '30 bis ∞']

    # Daten in Bins einteilen
    data['Temperature Bin'] = pd.cut(data['max.temp'], bins=temperature_bins, labels=temperature_labels, right=False)

    # Konvertiere 'datum' zu einem Datetime-Wert
    data['datum'] = pd.to_datetime(data['datum'], errors='coerce')

    # Erstelle Diagramme für jedes Jahr
    charts = []

    for year in [2019, 2020, 2021, 2022]:  # Ändere die Jahre entsprechend deiner Daten
        year_data = data[data['datum'].dt.year == year]
        chart = alt.Chart(year_data).mark_bar().encode(
            alt.Y('Temperature Bin:N', title='Temperaturbereich in °C', sort=temperature_labels),
            alt.X('gesamt:Q', title='Anzahl Radfahrer'),
            color=alt.Color('Temperature Bin:N', legend=alt.Legend(title='Temperaturbereiche'), sort=temperature_labels).scale(scheme="category10"),
            tooltip=['Temperature Bin:N', 'gesamt:Q']
        ).properties(
            title=f'{year}',
            width=300,  # Reduziere die Breite der Diagramme, um Platz für beide zu schaffen
            height=300
        )

        charts.append(chart)

    # Diagramme nebeneinander anzeigen
    combined_chart = alt.vconcat(
        alt.hconcat(charts[0], charts[1],
            title=alt.TitleParams(
            text='Anzahl Radfahrer pro Temperaturbereich von 2019 bis 2022',
            align='center',
            anchor='middle'
            ),       
                   ),
        alt.hconcat(charts[2], charts[3])
    ).resolve_scale(x='independent')

    return combined_chart

chart = create_charts(file_paths)

chart
```

# wie bei Korrelationen nach Datum gruppieren

```python
df_list = []

for file_path in file_paths:
    df_temp = pd.read_csv(file_path)
    df_list.append(df_temp)

# Einzelne DataFrames zu einem Gesamt-DataFrame zusammenführen
df = pd.concat(df_list, ignore_index=True)
```

```python
# Gruppiere den DataFrame nach 'datum' und summiere die Werte in den relevanten Spalten
df_summed = df.groupby('datum').agg({
    'gesamt': 'sum',
    'min.temp': 'mean',        # Durchschnitt der min.temp pro Tag
    'max.temp': 'mean',        # Durchschnitt der max.temp pro Tag
    'niederschlag': 'mean',
    'bewoelkung': 'mean',      # Durchschnitt der bewoelkung pro Tag
    'sonnenstunden': 'mean',
    'kommentar': 'first'       # Falls es Kommentare gibt, wird der erste genommen
}).reset_index()

# Ausgabe des resultierenden DataFrames
print(df_summed)
```

```python
def create_charts(file_paths):
    # Daten für alle Jahre kombinieren
    data = df_summed

    # Konvertiere 'niederschlag' zu numerischem Datentyp
    data['niederschlag'] = pd.to_numeric(data['niederschlag'], errors='coerce')

    # Bins für die Temperatur erstellen
    temperature_bins = [0, 5, 10, 20, float('inf')]
    temperature_labels = ['0 bis 5', '5 bis 10', '10 bis 20', '20 bis ∞']

    # Daten in Bins einteilen
    data['Niederschlag Bin'] = pd.cut(data['niederschlag'], bins=temperature_bins, labels=temperature_labels, right=False)

    # Konvertiere 'datum' zu einem Datetime-Wert
    data['datum'] = pd.to_datetime(data['datum'], errors='coerce')

    # Erstelle Diagramme für jedes Jahr
    charts = []

    for year in [2019, 2020, 2021, 2022]:  # Ändere die Jahre entsprechend deiner Daten
        year_data = data[data['datum'].dt.year == year]
        chart = alt.Chart(year_data).mark_bar().encode(
            alt.Y('Niederschlag Bin:N', title='Niederschlag in mm', sort=temperature_labels),
            alt.X('gesamt:Q', title='Anzahl Radfahrer'),
            color=alt.Color('Niederschlag Bin:N', legend=alt.Legend(title='Nierderschlagsbereiche'), sort=temperature_labels).scale(scheme="category10"),
            tooltip=['Niederschlag Bin:N', 'gesamt:Q']
        ).properties(
            title=f'{year}',
            width=300,  # Reduziere die Breite der Diagramme, um Platz für beide zu schaffen
            height=300
        )

        charts.append(chart)

    # Diagramme nebeneinander anzeigen
    combined_chart = alt.vconcat(
        alt.hconcat(charts[0], charts[1],
            title=alt.TitleParams(
            text='Anzahl Radfahrer pro Temperaturbereich von 2019 bis 2022',
            align='center',
            anchor='middle'
            ),       
                   ),
        alt.hconcat(charts[2], charts[3])
    ).resolve_scale(x='independent')

    return combined_chart

chart = create_charts(file_paths)

chart
```
